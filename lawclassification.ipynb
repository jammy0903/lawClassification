```python
# 필요한 라이브러리 설치
!pip install transformers
!pip install torch
!pip install gradio

from transformers import pipeline
import gradio as gr

class LegalClassifier:
    def __init__(self):
        # KLAID 모델 초기화
        self.classifier = pipeline(
            "text-classification",
            model="lawcompany/KLAID_LJP_base"
        )

    def classify_text(self, text):
        if not text:
            return "텍스트를 입력해주세요."

        # KLAID 모델로 분류 수행
        result = self.classifier(text)

        # 결과 포맷팅
        classification = result[0]
        label = classification['label']
        score = classification['score']

        return f"분류 결과: {label}\n신뢰도: {score:.4f}"

# Gradio 인터페이스 생성
def create_interface():
    classifier = LegalClassifier()

    iface = gr.Interface(
        fn=classifier.classify_text,
        inputs=gr.Textbox(label="법률 텍스트를 입력하세요"),
        outputs=gr.Textbox(label="분류 결과"),
        title="KLAID 법률 분류기",
        description="입력된 텍스트의 법률 분야를 분류합니다."
    )

    return iface

# 인터페이스 실행
demo = create_interface()
demo.launch(share=True)
```
